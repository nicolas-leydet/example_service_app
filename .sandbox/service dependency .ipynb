```python
class ServiceDependency(object):

    def setup(self):
        print('setup')

    def __get__(self, instance, cls):
        try:
            return self.__class__._lazy_instance
        except AttributeError:
            self.setup()
            self.__class__._lazy_instance = self
            return self

    def terminate(self):
        print('terminate')
        
    def do(self):
        print('do')
        
        
class Blue(object):
    db = ServiceDependency()
    
    def square(self):
        self.db.do()
        print('Blue square')
        
    def circle(self):
        self.db.do()
        print('Blue circle')
        

class Red(object):
    db = ServiceDependency()
    
    def square(self):
        self.db.do()
        print('Red square')
        
    def circle(self):
        self.db.do()
        print('Red circle')      
```

```python
b = Blue()
r = Red()
```
